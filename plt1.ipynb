```python
def pdf(x, mu, sigma):
    a = 1. / (sigma * np.sqrt(2.*np.pi))
    b = -1. / (2. * sigma**2)
    return a * np.exp(b * (x-mu)**2)

x = np.linspace(-6, 6, 1000)
for i in range(5):
    data = np.random.standard_normal(50)
    mu, sigma = np.mean(data), np.std(data)
    plt.plot(x, pdf(x, mu, sigma), color='.75')
    
plt.plot(x, pdf(x, 0, 1), color='k') 
plt.show()
```

```python
def pdf(x, mu, sigma):
    a = 1. / (sigma * np.sqrt(2.*np.pi))
    b = -1. / (2. * sigma**2)
    return a * np.exp(b * (x-mu)**2)

x = np.linspace(-10, 10, 1000)
for i in np.arange(1, 4, .2):
    plt.plot(x, pdf(x, 0, i), color='b')
    
plt.plot(x, pdf(x, 0, 1), color='k') 
plt.show()
```

```python
# 对scatter plot使用颜色
import numpy as np
import matplotlib.pyplot as plt

A = np.random.standard_normal((100, 2))
A += np.array([-1, -1]) # 使随机点分布于(-1, -1)
B = np.random.standard_normal((100, 2))
B += np.array([1, 1]) # 使随机点分布于(1, 1)

plt.scatter(A[:, 0], A[:, 1], color='0.25')
plt.scatter(B[:, 0], B[:, 1], color='0.75')
plt.scatter(1, 1, color='g')
plt.scatter(-1, -1, color='b')
plt.show()
```

```python
# 下列例子演示Numpy函数loadtxt()的使用
# 为了方便演示，使用StringIO来实时(on the fly)创建文件对象
from io import StringIO   # StringIO behaves like a file object
import numpy as np

# Exam 1:一个简单例子，文件中的数字由空格分开
f = StringIO('1 2 3\n4 5 6')
data = np.loadtxt(f)
print('Exam 1:')
print(data)

# Exam 2:文件中的数字由逗号','分开
f = StringIO('1,2,3\n4,5,6')
data = np.loadtxt(f, delimiter=',')
print('Exam 2:')
print(data)

# Exam 2+:文件中的数字由逗号','分开；但是在逗号后面故意留有空格
f = StringIO('1, 2, 3\n4, 5, 6')
data = np.loadtxt(f, delimiter=',') # 只要指定'delimiter'为','就可以，空格不影响对数字的解析
print('Exam 2+:')
print(data)

# Exam 3: 只使用文件中指定的列
f = StringIO('1, 2, 3\n4, 5, 6')
data = np.loadtxt(f, delimiter=',', usecols=(0, 2))
print('Exam 3:')
print(data)

# Exam 4: 在文件中包含注释
f = StringIO('1, 2, 3 # line 1\n4, 5, 6 # line 2') # 使用'#'作为注释开始的符号
data = np.loadtxt(f, delimiter=',') # 'loadtxt()'函数默认识别'#'为注释符号，其在本行内后面的内容都被认为是注释
print('Exam 4:')
print(data)

# Exam 4+: 在文件中包含注释
f = StringIO('1, 2, 3 -- line 1\n4, 5, 6 -- line 2') # 使用'--'作为注释开始的符号
data = np.loadtxt(f, delimiter=',', comments='--')
print('Exam 4+:')
print(data)

# Exam 4++: 在文件中包含注释
f = StringIO('1, 2, 3 // line 1\n4, 5, 6 // line 2') # 使用'//'作为注释开始的符号
data = np.loadtxt(f, delimiter=',', comments='//')
print('Exam 4++:')
print(data)

# Exam 5: 跳过文件开头的几行
f = StringIO('this is first line.\nthis is second line: num1, num2, num3\n1, 2, 3\n4, 5, 6')
data = np.loadtxt(f, delimiter=',', skiprows=2) # 使用'skiprows'参数跳过前2行
print('Exam 5:')
print(data)

# Exam 6: 按‘列’返回文件内容
f = StringIO('1, 2, 3 # line 1\n4, 5, 6 # line 2')
col0, col2 = np.loadtxt(f, delimiter=',', usecols=(0, 2), unpack=True) # 使用'unpack'参数按列返回内容
print('Exam 6:')
print('0 column: {}'.format(col0))
print('2 column: {}'.format(col2))

# Exam 7a: 使用'converters'来‘手动’解析某些列
f = StringIO('1, 2, 3 # line 1\n4, 5,  # line 2') # 注意：第二行第三列的值缺失
data = np.loadtxt(f, delimiter=',', converters={2: lambda s:float(s.strip() or -1)}) # 使用'-1'来填充缺失的数值
print('Exam 7a:')
print(data)

# Exam 7b: 将字符串映射到数值
f = StringIO('1, 2, 3, red # line 1\n4, 5, 6, blue # line 2') # 新增第四列为颜色名称
colors = [b'red', b'green', b'blue'] # 颜色列表，用于将颜色名称映射到列表索引
data = np.loadtxt(f, delimiter=',', converters={3: lambda s:colors.index(s.strip())})
print('Exam 7b:')
print(data)

# Exam 7b+: converters里面也可以使用函数
def color_index(c):
    return colors.index(c.strip())

f = StringIO('1, 2, 3, red # line 1\n4, 5, 6, blue # line 2') # 新增第四列为颜色名称
data = np.loadtxt(f, delimiter=',', converters={3: color_index})
print('Exam 7b+:')
print(data)

# Exam 8: 使用'ndmin'来指定返回的ndarray数组的最小维度
f = StringIO('1, 2, 3 # line 1\n\n') # 注意文件中只包含一行数据（文件末尾连续的换行被忽略）
data = np.loadtxt(f, delimiter=',', ndmin=2) # 将返回的数组维持在2维。注意'ndmin'只能取值为0, 1, 2，默认为0
print('Exam 8:')
print(data)

# Exam 8+: 将返回值维持在2维
f = StringIO('100')
data = np.loadtxt(f, ndmin=2)
print('Exam 8+:')
print(data)
```
