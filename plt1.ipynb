```python
# Python列表相加和NumPy数组相加的区别
A = [5, 30, 45, 22]
B = [5, 25, 50, 20]
C = A + B
print('Python list addition: {}'.format(C))
A = np.array([5, 30, 45, 22])
B = np.array([5, 25, 50, 20])
C = A + B
print('NumPy array addition: {}'.format(C))
```

```python
# stacked bar chart, extended example
data = np.random.rand(5, 4)
x = np.arange(data.shape[1])
colors = ['r', 'g', 'b']

sum = np.zeros(data.shape[1])
for i, row in enumerate(data):
    plt.bar(x, row, color=colors[i%len(colors)], bottom=sum)
    sum += row
    
plt.show()
```

```python
def pdf(x, mu, sigma):
    a = 1. / (sigma * np.sqrt(2.*np.pi))
    b = -1. / (2. * sigma**2)
    return a * np.exp(b * (x-mu)**2)

x = np.linspace(-6, 6, 1000)
for i in range(5):
    data = np.random.standard_normal(50)
    mu, sigma = np.mean(data), np.std(data)
    plt.plot(x, pdf(x, mu, sigma), color='.75')
    
plt.plot(x, pdf(x, 0, 1), color='k') 
plt.show()
```

```python
def pdf(x, mu, sigma):
    a = 1. / (sigma * np.sqrt(2.*np.pi))
    b = -1. / (2. * sigma**2)
    return a * np.exp(b * (x-mu)**2)

x = np.linspace(-10, 10, 1000)
for i in np.arange(1, 4, .2):
    plt.plot(x, pdf(x, 0, i), color='b')
    
plt.plot(x, pdf(x, 0, 1), color='k') 
plt.show()
```

```python
# 对scatter plot使用颜色
import numpy as np
import matplotlib.pyplot as plt

A = np.random.standard_normal((100, 2))
A += np.array([-1, -1]) # 使随机点分布于(-1, -1)
B = np.random.standard_normal((100, 2))
B += np.array([1, 1]) # 使随机点分布于(1, 1)

plt.scatter(A[:, 0], A[:, 1], color='0.25')
plt.scatter(B[:, 0], B[:, 1], color='0.75')
plt.scatter(1, 1, color='g')
plt.scatter(-1, -1, color='b')
plt.show()
```

```python
# 这个例子演示Numpy函数loadtxt()的使用
# 为了方便演示，使用StringIO来实时(on the fly)创建文件对象
from io import StringIO   # StringIO behaves like a file object

# Exam 1:一个简单例子，文件中的数字由空格分开
f = StringIO('1 2 3\n4 5 6')
data = np.loadtxt(f)
print('Exam 1:')
print(data)

# Exam 2:文件中的数字由逗号','分开
f = StringIO('1,2,3\n4,5,6')
data = np.loadtxt(f, delimiter=',')
print('Exam 2:')
print(data)

# Exam 2+:文件中的数字由逗号','分开；但是在逗号后面故意留有空格
f = StringIO('1, 2, 3\n4, 5, 6')
data = np.loadtxt(f, delimiter=',') # 只要指定'delimiter'为','就可以，空格不影响对数字的解析
print('Exam 2+:')
print(data)

# Exam 3: 只使用文件中指定的列
f = StringIO('1, 2, 3\n4, 5, 6')
data = np.loadtxt(f, delimiter=',', usecols=(0, 2))
print('Exam 3:')
print(data)

# Exam 4: 在文件中包含注释
f = StringIO('1, 2, 3 # line 1\n4, 5, 6 # line 2') # 使用'#'作为注释开始的符号
data = np.loadtxt(f, delimiter=',') # 'loadtxt()'函数默认识别'#'为注释符号，其在本行内后面的内容都被认为是注释
print('Exam 4:')
print(data)

# Exam 4+: 在文件中包含注释
f = StringIO('1, 2, 3 -- line 1\n4, 5, 6 -- line 2') # 使用'--'作为注释开始的符号
data = np.loadtxt(f, delimiter=',', comments='--')
print('Exam 4+:')
print(data)

# Exam 4++: 在文件中包含注释
f = StringIO('1, 2, 3 // line 1\n4, 5, 6 // line 2') # 使用'//'作为注释开始的符号
data = np.loadtxt(f, delimiter=',', comments='//')
print('Exam 4++:')
print(data)
```
